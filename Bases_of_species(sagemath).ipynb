```python
import pandas as pd
f = pd.read_pickle("~/species_of_xmls_for_m2.csv")
#f.values[i][0]:parameter f.values[i][1]:species f.values[i][2]:formulas
```

```python
#
#p:parameter s:species fo:formulas name:modelname
bases=[]
for i in range(len(f)):
    p,s,fo,name=f.values[i][0],f.values[i][1],f.values[i][2],f.values[i][3]
    
    #
    if 0<len(f.values[i][2])<=15:
        
        #m
        ps=""
        for j in range(len(p)):
            if j<len(p)-1:
                ps+=p[j]+"=random(QQ),"
            else:
                ps+=p[j]+"=random(QQ)"
        
        ss="species = {"
        for j in range(len(s)):
            if j<len(s)-1:
                ss+=s[j]+","
            else:
                ss+=s[j]+"}"

        fos="model = try ideal("
        for j in range(len(fo)):
            if j<len(fo)-1:
                fos+=fo[j]+","
            else:
                fos+=fo[j]+") else 0"

        #formulapow(a,b) a^b
        k=0
        while fos.find("pow",k+1)!=-1:
            k=fos.find("pow",k)

            #pow
            kk=k+3
            ki=1
            while ki!=0:
                kk+=1
                if fos[kk]=="(":
                    ki+=1
                elif fos[kk]==")":
                    ki-=1
                elif fos[kk]==",":
                    kc=kk
            #kpowkckk
            fos=fos[:k]+"("+fos[k+4:kc]+")^("+fos[kc+1:kk]+")"+fos[kk+1:]

        #_     
        ps = ps.replace("_", "u")
        ss = ss.replace("_", "u")
        fos = fos.replace("_", "u")
        
        ps = ps.replace("random", "rando")
        ps = ps.replace("do", "dom")
        ss = ss.replace("do", "dom")
        fos = fos.replace("do", "dom")
        
        
        fos = fos.replace("numerator", "numerato")
        ps = ps.replace("to", "tor")
        ss = ss.replace("to", "tor")
        fos = fos.replace("to", "tor")
        
        def remo(ps,ss,fos,st):
            st2=st+st[-1]*3
            ps = ps.replace(st, st2)
            ss = ss.replace(st, st2)
            fos = fos.replace(st, st2)
            return(ps,ss,fos)
        
        ps,ss,fos=remo(ps,ss,fos,"R")
        ps,ss,fos=remo(ps,ss,fos,"Sum")
        ps,ss,fos=remo(ps,ss,fos,"sum")
        ps,ss,fos=remo(ps,ss,fos,"pi")
        ps,ss,fos=remo(ps,ss,fos,"chi")
        ps,ss,fos=remo(ps,ss,fos,"Gamma")
        ps,ss,fos=remo(ps,ss,fos,"log")
        ps,ss,fos=remo(ps,ss,fos,"zeta")
        ps,ss,fos=remo(ps,ss,fos,"of")
        ps,ss,fos=remo(ps,ss,fos,"step")
        
        
        #
        if "exp" in fos:
            continue
        if "and" in fos:
            continue
        if "leq" in fos:
            continue

        #
        if  i==303 or i==314: 
            continue
        
        #m2
        macaulay2("restart")
        macaulay2('load("~/Documents/matroids.m2")')

        print(ps)
        macaulay2(ps)
        print("psOK")
        
        print(ss)
        macaulay2(ss)
        print("ssOK")
        macaulay2("R = QQ[species]")
        print(fos)
        macaulay2(fos)
        print("fosOK")
        
        print(i , name)
        print('load("~/Documents/matroids.m2")')
        print(ps)
        print(ss)
        print("R = QQ[species]")
        print(fos)
        if macaulay2("model") != 0:
            if macaulay2("circuits(R,model)")!=0:
                try:
                    bases.append([name, ss, macaulay2("circuits(R,model)").sage()])
                except NotImplementedError: #ideal()
                    pass
```

```python
bbb=bases
l=[]
for i in range(len(bbb)):
    if bbb[i][2]!=[[]] and bbb[i][2]!=[]:
        l.append(bbb[i])
```
